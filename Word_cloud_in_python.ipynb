```python
import pandas as pd

data=pd.read_csv("IMDB-Dataset.csv")

#understanding the dataset and their columns name we work on review col in which we extract most frequent words
# print(data.columns)
# print(data['review'][0])

#Data Cleaning
#remove punctuation
#convert to lowercase
#remove stopwords like the,is,and
#re.sub():Remove punctuation and STOPWORD-set of stopwords

import re
from wordcloud import STOPWORDS

text = ' '.join(data['review'].astype(str).tolist())

text = re.sub(r'[^A-Za-z\s]', '', text)

text = text.lower()

stopwords = set(STOPWORDS)
text = ' '.join(word for word in text.split() if word not in stopwords)

#generating word cloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud=WordCloud(width=1200,height=600,background_color='white').generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.title('IMDB Movie Review Word Cloud')
plt.show()
```

```python
wordcloud.to_file("wordcloud.png")
```
